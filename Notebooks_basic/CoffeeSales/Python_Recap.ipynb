# 11. Joining Tables
Two tables. One shared column. Combine them.

> *"In real projects data is never in one table. Customer info lives in one table, orders in another, products in another. Joining is how you bring them together."*

**Why use an ID to join?**
Because names can repeat. Two students can be called Alice. But `student_id = 101` is unique — it belongs to exactly one person. Real databases always join on IDs, never on names.

## The three tables we have

````python
# Table 1 — students (already exists from the setup cell)
print("Table 1 — students:")
print(students[["student_id", "name", "grade"]].head())
print()

# Table 2 — contact info
contacts = pd.DataFrame({
    "student_id": [101, 102, 103, 104, 105,
                   106, 107, 108, 109, 110],
    "email": ["alice@uni.de",  "bob@uni.de",   "carol@uni.de",
              "david@uni.de",  "eva@uni.de",   "frank@uni.de",
              "grace@uni.de",  "henry@uni.de", "iris@uni.de",
              "james@uni.de"]
})
print("Table 2 — contacts:")
print(contacts.head())
print()

# Table 3 — course registrations (only 5 students registered)
courses = pd.DataFrame({
    "student_id": [101, 103, 105, 107, 109],
    "course":     ["Data Science", "Machine Learning",
                   "Data Science", "Statistics", "Machine Learning"]
})
print("Table 3 — course registrations (only 5 students):")
print(courses)
````

## The basic merge — combining two tables
Pattern:
```python
table1.merge(table2, on="shared_column")
```
This is called an **inner join** by default.
It keeps only rows that exist in **both** tables.

````python
# Merge students with contacts using student_id
students_full = students.merge(contacts, on="student_id")

print("Result — students with email:")
print(students_full[["student_id", "name", "grade", "email"]])
print()
print("Rows:", len(students_full))
# All 10 students — because every student has a contact row
````

## The four types of join
This is what makes joining powerful — you control which rows to keep.

| Join type | What it keeps | How to write it |
|---|---|---|
| Inner join | Only rows in **both** tables | `how="inner"` (default) |
| Left join | **All** rows from the left table | `how="left"` |
| Right join | **All** rows from the right table | `how="right"` |
| Outer join | **All** rows from **both** tables | `how="outer"` |

> *"Think of two circles overlapping. Inner = only the overlap. Left = the whole left circle. Outer = both circles completely."*

## Inner join — only matching rows

````python
# Inner join — students merged with courses
# Only students who registered for a course appear
inner = students.merge(courses, on="student_id", how="inner")

print("Inner join result:")
print(inner[["name", "grade", "course"]])
print()
print("Rows:", len(inner), "— only the 5 students who registered")
````

## Left join — keep all rows from the left table

````python
# Left join — all students appear
# Students who did not register get NaN in the course column
left = students.merge(courses, on="student_id", how="left")

print("Left join result:")
print(left[["name", "grade", "course"]])
print()
print("Rows:", len(left), "— all 10 students")
print()
print("NaN means: this student did not register for any course")
````

## Inner vs Left — the key difference
Run both and compare the output:
- **Inner:** 5 rows — only registered students
- **Left:** 10 rows — all students, NaN for those not registered

> *"Left join is safer for analysis. If you use inner join and someone has no course — they disappear from your data silently. Left join keeps them visible with NaN so you know they exist."*

````python
# Side by side comparison
print("INNER JOIN — rows:", len(inner))
print(inner[["name", "course"]])
print()
print("LEFT JOIN — rows:", len(left))
print(left[["name", "course"]])
````

## Outer join — keep everything from both tables

````python
# Add a student to courses that does NOT exist in students
courses_extra = pd.DataFrame({
    "student_id": [101, 103, 105, 107, 109, 999],  # 999 does not exist
    "course":     ["Data Science", "Machine Learning",
                   "Data Science", "Statistics",
                   "Machine Learning", "Python Basics"]
})

# Outer join — keeps all rows from BOTH tables
outer = students.merge(courses_extra, on="student_id", how="outer")

print("Outer join result:")
print(outer[["student_id", "name", "course"]])
print()
print("Notice student_id 999 — exists in courses but not in students")
print("Name is NaN because we have no student info for that ID")
````

## When to use which join

| Situation | Use |
|---|---|
| You only want complete records | Inner join |
| You want all students even if some data is missing | Left join |
| You want everything from both tables | Outer join |
| You want all records from the second table | Right join |

**Left join is the most common in data analysis.** It keeps your main table intact and adds extra information where available.
